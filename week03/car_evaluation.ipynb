```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
## index_col = 0 : 열 인덱스를 0,1,2,3, 이렇게 인덱스 부여
# header = None: header를 0,1,2,3,4,... 이렇게 인덱스 부여
df = pd.read_csv("/content/drive/MyDrive/car_evaluation.csv", header = None)
df
```

```python
columns = ['buying','maint','doors','persons','lug_boot','safety','class']
```

```python
label_encoders = {}
for column in columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
```

```python
# 데이터 확인
df['class'].value_counts()
```

```python
X = df.drop('class', axis=1).values
y = df['class']
```

```python
# X 데이터를 표준화한다. 즉 각 특정의 평균을 0, 표준편차를 1로 변환하는 역할을 한다.
scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# DT (걸정 트리)
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
dt_accuracy = accuracy_score(y_test, y_pred)
dt_cm = confusion_matrix(y_test, y_pred)
print(f'DT 정확도 {dt_accuracy:.2f}')
print(dt_cm)
```

```python
# RF
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
rf_accuracy = accuracy_score(y_test, y_pred)
rf_cm = confusion_matrix(y_test, y_pred)
print(f'RF 정확도 {rf_accuracy:.2f}')
print(rf_cm)
```

```python
# SVM
svm = SVC()
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
svm_accuracy = accuracy_score(y_test, y_pred)
svm_cm = confusion_matrix(y_test, y_pred)
print(f'SVM 정확도 {svm_accuracy:.2f}')
print(svm_cm)
```

```python
# LR
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
lr_accuracy = accuracy_score(y_test, y_pred)
lr_cm = confusion_matrix(y_test, y_pred)
print(f'lr 정확도 {lr_accuracy:.2f}')
print(lr_cm)
```

```python
# KNN
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
knn_accuracy = accuracy_score(y_test, y_pred)
knn_cm = confusion_matrix(y_test, y_pred)
print(f'knn 정확도 {knn_accuracy:.2f}')
print(knn_cm)
```
