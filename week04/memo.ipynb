# UnderFitting과 Overfitting
- unserFitting은 학습이 덜됐다
- OverFitting은 학습이 너무 많이 됐다.
- 학습이란 룰을 습득하는 것임

## 모델의 일반화
- 모든 모델은 일반화 능력을 가져야함
- 다른 데이터도 처리 가능해야함
- 예시로 백인의 피부데이터를 가지고 우리나라에서 치료하면 잘 안맞음 -> 일반화가 안됨

|               | Underfitting               | overfitting                               |
|---------------|----------------------------|-------------------------------------------|
| what(정의)      | 학습이 덜 된 상태                 | 학습이 너무 잘된 상태                              |
| why(왜 발생함)    | Data 부족 -> 일반화가 안된다        | 데이터가 너무 복잡하게 되어있음 -> noise까지도 학습 |
| solotion(해결법) | 데이터 증대기법 사용 -> 데이터를 늘리면 된다. | 모델을 심플하게 바꾸어야한다.        |

### 회귀
- x로부터 y의 상관관계를 구하는게 회귀임
- R^2을 사용해서 상관관계를 명확하게 할 수 있다.
- R^2 계산은 책 121페이지 참고  계산 방법 시험에는 안나옴
- 즉, 상관관계는 두 변수 간에 서로 얼마나 연관있는지 나타내는 지표

### 특성 공학과 규제
#### 피처 엔지니어링 ( 중요 )
- 새롭게 만들거나 기존의 있는 특성으로 만드는 것을 말함
- 중요한건 새롭게 만든 피처가 N하고 상관관계가 적었을 때, 도움이 된다.
- 특성이 있는 애들만 뽑아서 가는것을 말함

### 규제( P.163 )
#### 규제가 필요한 이유(목적)
- overfitting 방지 즉, 모델이 외우지 않고 일반화하게 만드는 장치 다시말해, 새로운 데이터에도 잘 작동하도록 도와주는 역할
- Norm1(L1, 라쏘) , Norm2(L2, 릿지)

### RF
- DT를 여러개 사용해서 제일 좋은 것을 선택
-  RF = Bootstrap + Decision Tree + Random Feature Selection
- 위에 것들중 하나라도 없으면 RF가 성립 안함

#### 앙상블
- 여러 모델을 모아서 제일 좋은 모델을 사용하는 것
- 동시에 여러개를 사용하니 오버헤드가 크다. 하드웨어 자원을 많이 사용함

#### RF가 앙상블인 이유 (중요)
- DT를 여러개 사용해서 좋은것을 사용하니까

#### bootstrap
- 데이터 세트에서 중복을 허용하여 데이터를 샘플링하는 기법
- 1~5까지의 중복 샘플을 만듬
- 즉, 1,1,2,3,4 또는 1,2,2,3,4 이런식으로 여러개의 샘플을 만듬
- RF 방식
    - 1,1,2,3,4 이런거에 DT를 붙임 -> 이런것들중에 제일 좋은것을 선택 그것이 RF
